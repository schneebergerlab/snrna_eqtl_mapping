```python
def dist_to_eqtl(eqtl_df, bin_size=25000):
    is_right_of_eqtl = eqtl_df.pos_gene > eqtl_df.pos_eqtl
    is_exact = (eqtl_df.pos_eqtl <= eqtl_df.pos_gene) & (eqtl_df.pos_gene < (eqtl_df.pos_eqtl + bin_size))
    dist = eqtl_df.pos_eqtl - eqtl_df.pos_gene
    dist[is_right_of_eqtl] = eqtl_df[is_right_of_eqtl].pos_gene - (eqtl_df[is_right_of_eqtl].pos_eqtl + bin_size)
    dist[is_exact] = 0
    return dist
```
