```python
def bootstrap_conf_interval(arr, percentiles=[2.5, 97.5], nboot=1000, estimator=np.median):
    boots = []
    for _ in range(nboot):
        boots.append(estimator(np.random.choice(arr, size=len(arr), replace=True)))
    return np.percentile(boots, percentiles)

nmarkers = 10 ** cb_stats.query('cluster == "hq"').nmarkers.values
bootstrap_conf_interval(nmarkers)
```

```python
hap_probs = pd.concat((
    pd.read_csv(fn, sep='\t', header=[0, 1], index_col=[0, 1])
    for fn in snakemake.input.crossovers
), axis=0)
hap_probs.index = hap_probs.index.droplevel('fn_idx')
hap_probs.columns = hap_probs.columns.set_levels(hap_probs.columns.levels[1].astype(int), level=1)
```

```python
chrom_sizes = get_chrom_sizes_bam(snakemake.input.bams[0])
```

```python
def calculate_cm_per_mb(hap_probs, nboots=100, bin_size=25_000, cm_bin_size=1_000_000, min_prob=0.01):

    nf = 1_000_000 // bin_size
    ws = cm_bin_size // bin_size
    mw = ws // 2
    
    boots = {}
    for chrom, chrom_hap_probs in hap_probs.groupby(level=0, axis=1, group_keys=False):
        chrom_hap_probs.columns = chrom_hap_probs.columns.droplevel(0)
        chrom_co_probs = (chrom_hap_probs.diff(1, axis=1)
                                         .abs()
                                         .fillna(0))
        chrom_co_probs.mask(chrom_co_probs < min_prob, other=0, inplace=True)
        # equivalent to rolling sum accounting for edge effects
        chrom_co_probs = chrom_co_probs.rolling(ws, center=True, min_periods=mw, axis=1).mean() * nf
        chrom_boots = []
        for _ in range(nboots):
            cm_samp = chrom_co_probs.sample(len(chrom_co_probs), replace=True).mean(axis=0)
            chrom_boots.append(cm_samp * 100)
        boots[chrom] = pd.concat(chrom_boots, axis=1)
    return boots
```

```python
idx = cb_stats.query('cluster == "hq"').cb
n = len(idx)
hp = hap_probs.loc[idx]
cm = calculate_cm_per_mb(hp, nboots=100, cm_bin_size=100_000)
tot_co = hp.groupby(level=0, axis=1).diff(1, axis=1).abs().sum().sum()
fig, axes = chrom_axes_boilerplate(chrom_sizes, figsize=(12, 3))
axes = axes.flatten()
for chrom, ax in zip(chrom_sizes, axes):
    ax.step(cm[chrom].index.astype(int), cm[chrom].mean(1).values, color=c)
    ax.fill_between(x=cm[chrom].index.astype(int), y1=np.percentile(cm[chrom].values, 2.5, axis=1), y2=np.percentile(cm[chrom].values, 97.5, axis=1), alpha=0.25, color=c)
axes[0].set_ylabel('cM / Mb')
axes[0].set_ylim(-1, 12.5)
fig.suptitle(f'Col-0 x Db-1 crossover landscape (n = {n} nuclei, {tot_co:.0f} COs)')
plt.tight_layout()
plt.savefig(os.path.join(snakemake.output[0], f'co_landscape.svg'))
plt.show()
```
