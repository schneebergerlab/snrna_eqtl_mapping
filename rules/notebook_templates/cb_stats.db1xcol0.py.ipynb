```python
def read_cb_stats(geno_fn, co_stats_fns, hap_probs_fns, haplotype_uncertainty_percentile=75):
    if geno_fn is not None:
        geno = pd.read_csv(
            geno_fn,
            sep='\t',
            header=0,
            names=['cb', 'geno', 'nsnps', 'geno_logprob'],
            index_col='cb',
        )
        geno['nsnps'] = np.log10(geno.nsnps)
    else:
        geno = None
    co = []
    for fn in co_stats_fns:
        co.append(
            pd.read_csv(fn,
                        sep='\t',
                        index_col=['cb'],
                        usecols=[1, 2, 3],
                        names=['cb', 'nmarkers', 'hmm_score'],
                        header=0)
            )
    co = pd.concat(co, axis=0)
    co['nmarkers'] = np.log10(co.nmarkers)
    co['hmm_score'] = np.minimum(-np.log2(1 - co.hmm_score), 10)

    hap_probs = pd.concat((
        pd.read_csv(fn, sep='\t', header=[0, 1], index_col=[0, 1])
        for fn in hap_probs_fns
    ), axis=0)
    hap_probs.index = hap_probs.index.droplevel('fn_idx')

    # Since ambient DNA or doublet contamination is more likely to come from Col-0, 
    # we can use mean haplotype as a filtering metric
    co['mean_hap'] = hap_probs.mean(axis=1)

    # use the area under the absolute uncertainty line to estimate overall uncertainty
    hap_uncertainty = (hap_probs - (hap_probs > 0.5)).abs()
    co['hap_uncertainty'] = np.maximum(
        np.log10(hap_uncertainty.groupby(level=0, axis=1).agg(np.trapz, axis=1).sum(axis=1)),
        0
    )

    co_probs = (hap_probs.groupby(level=0, axis=1)
                         .diff(1, axis=1)
                         .abs()
                         .dropna(axis=1))
    co['n_crossovers'] = co_probs.sum(axis=1)

    return geno.join(co) if geno is not None else co, hap_probs
```

```python

USESTATS = [
    'hmm_score', 'hap_uncertainty', 'mean_hap', 'nmarkers',
]

def cluster_hq_barcodes(barcode_stats, init_nmarker_threshold=2, stat_names=USESTATS):
    # good cells should in general have high geno probs
    init_label = barcode_stats.eval('(hmm_score > 4 | hap_uncertainty < 2.75) & nmarkers > @init_nmarker_threshold').astype(int)
    print(init_label.value_counts())
    barcode_stats = barcode_stats.copy()[stat_names]
    dists = []
    for i in (0, 1):
        X_i = barcode_stats.loc[init_label == i]
        dists.append(pm.IndependentComponentsDistribution([
            pm.NormalDistribution(X_i[sn].mean(), X_i[sn].std())
            for sn in stat_names
        ]))
    _, w = np.unique(init_label, return_counts=True)
    gmm = pm.GeneralMixtureModel(dists, weights= w / w.sum())
    gmm.fit(barcode_stats.values, inertia=0.9)
    pred = gmm.predict(barcode_stats.values)
    prob = gmm.predict_proba(barcode_stats.values)[:, 1]
    prob = pd.Series(prob, index=barcode_stats.index, name='prob_hq')
    return prob
```

```python
cb_stats['prob_hq'] = cluster_hq_barcodes(cb_stats)
```
