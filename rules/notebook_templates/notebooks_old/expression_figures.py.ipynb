```python
def read_index(idx_fn):
    with open(idx_fn) as f:
        idx = [r.strip().split('\t')[0] for r in f.readlines()]
    return np.array(idx)


def lognormalise_exprs(mtx, alpha=0.25):
    umi_counts = mtx.sum(0)
    sf = (umi_counts / np.mean(umi_counts))
    ln_mtx = np.log2(mtx / sf + (1.0 / (4 * alpha)))
    sf = sf.rename('size_factors')
    return ln_mtx, sf


def read_mmformat(mtx_dirs, cb_whitelist, feat_filt=10,
                  mtx_fn='matrix.mtx',
                  barcode_fn='barcodes.tsv',
                  feature_fn='features.tsv'):
    mtxs = []
    for mtx_dir in mtx_dirs:
        mtx = mmread(os.path.join(mtx_dir, mtx_fn)).tocsr()
        bc_mask = np.array(mtx.sum(axis=0) > 0).ravel()
        mtx = mtx[:, bc_mask]
        barcodes = read_index(os.path.join(mtx_dir, barcode_fn))[bc_mask]
        features = read_index(os.path.join(mtx_dir, feature_fn))
        mtx = pd.DataFrame.sparse.from_spmatrix(mtx, columns=barcodes, index=features)
        mtx = mtx.loc[:, mtx.columns.isin(cb_whitelist)]
        mtxs.append(mtx)
    mtx = pd.concat(mtxs, axis=1).fillna(0)
    #mtx = mtx.loc[np.count_nonzero(mtx, axis=1) > feat_filt]
    mtx = mtx.sparse.to_dense().astype(np.float32)
    return mtx
```

```python
cb_stats = pd.read_csv(
    'cb_stats/col0xmany.tsv',
    header=0,
    index_col=0,
    sep='\t',
    usecols=['cb', 'cluster'],
).query('cluster == "hq"')

cb_whitelist = sorted(cb_stats.index.tolist())

snakemake.input.solo_output = glob('aligned_data/diploid/col0xmany.col0__vs__*_genecounts/')
count_mat = read_mmformat(
    snakemake.input.solo_output,
    cb_whitelist
)
exprs_mat, sf = lognormalise_exprs(count_mat)
cb_stats['sequencing_depth'] = np.log10(sf)
cb_stats['n_genes_exprs'] = np.log2((exprs_mat > 0).sum(0))
```

```python
pca = PCA(n_components=10, svd_solver='arpack')
pca_proj = pca.fit_transform(exprs_mat.T)
pca_proj = pd.DataFrame(pca_proj, index=exprs_mat.columns, columns=[f'PC{i}' for i in range(1, 11)])

cluster_labels = pd.Series(GaussianMixture(covariance_type='spherical', n_components=2, random_state=1).fit(pca_proj).predict(pca_proj), index=pca_proj.index)
cb_stats['celltype_cluster'] = cluster_labels
clusters, cluster_counts = np.unique(cluster_labels, return_counts=True)


sns.jointplot(
    x=pca_proj['PC1'],
    y=pca_proj['PC2'],
    hue=(cluster_labels > 0.5).astype(int).map(lambda x: f'Cluster {x} (n={cluster_counts[x]})'),
    marginal_kws={'common_norm': False},
    joint_kws={'rasterized': True},
    palette=pal[:2],
)

plt.tight_layout()
#plt.savefig(os.path.join(snakemake.output[0], 'pca_cluster_jointplot.svg'), dpi=500)
plt.show()
```

```python
agg_by_ref = exprs_mat.groupby(exprs_mat.index.str[:3]).sum()
cb_stats['percent_mitochondrial'] = agg_by_ref.loc['ATM'] / agg_by_ref.sum(0) * 100
cb_stats['percent_chloroplast'] = agg_by_ref.loc['ATC'] / agg_by_ref.sum(0) * 100
```
