```python
import numpy as np
from scipy.signal import find_peaks
from scipy.ndimage import convolve
import pandas as pd

LOD_THRESHOLD = 3.0
FDR_THRESHOLD = 0.05

def find_lod_peaks(eqtl_data, lod_threshold=LOD_THRESHOLD, rel_lod_threshold=0.1,
                   fdr_threshold=FDR_THRESHOLD,
                   rel_prominence=0.25, lod_drop=1.5,
                   min_dist=3_000_000, bin_size=25_000,
                   cell_type_specific=False):
    min_dist = min_dist // bin_size
    peaks = []
    for _, chrom_eqtl_data in eqtl_data.groupby('chrom'):
        chrom_eqtl_data = chrom_eqtl_data.sort_values('pos')
        lod = chrom_eqtl_data.lod_score.values if not cell_type_specific else np.maximum(chrom_eqtl_data.cluster_lod_score.values, chrom_eqtl_data.lod_score.values)
        lod[lod < lod_threshold] = -np.inf
        fdr = chrom_eqtl_data.lrt_pval.values if not cell_type_specific else np.minimum(chrom_eqtl_data.cluster_lrt_pval.values, chrom_eqtl_data.lrt_pval.values) * 2
        peak_idx, _ = find_peaks(
            np.insert(lod, [0, len(lod)], [-np.inf, -np.inf]), # insert infs at beginning and end of chromosome to prevent prominence edge effects
            height=max(lod_threshold, max(lod) * rel_lod_threshold),
            prominence=max(lod) * rel_prominence,
            distance=min_dist
        )
        peak_idx = peak_idx - 1
        peak_idx = [i for i in peak_idx if fdr[i] < fdr_threshold]
        l_ci_idx = 0
        r_ci_idx = len(chrom_eqtl_data) - 1
        for p_idx in peak_idx:
            drop_thresh = lod[p_idx] - lod_drop
            for r_ci_idx in range(p_idx, len(lod)):
                if lod[r_ci_idx] <= drop_thresh:
                    break
            for l_ci_idx in reversed(range(0, p_idx)):
                if lod[l_ci_idx] <= drop_thresh:
                    break
            eqtl_peak = chrom_eqtl_data.iloc[p_idx].copy()
            eqtl_peak['start'] = chrom_eqtl_data.iloc[l_ci_idx].pos
            eqtl_peak['end'] = chrom_eqtl_data.iloc[r_ci_idx].pos + bin_size
            peaks.append(eqtl_peak)
    peaks = pd.DataFrame(
        peaks,
        columns=eqtl_data.columns.tolist() + ['start', 'end'],
    )
    peaks['pos_eqtl'] = peaks.pop('pos').astype(int)
    peaks['start'] = peaks.start.astype(int)
    peaks['end'] = peaks.end.astype(int)
    return peaks
```

```python
sig_eqtl = pd.merge(
    eqtl_res.groupby('gene_id', as_index=False).apply(find_lod_peaks).reset_index(drop=True),
    gene_locs,
    left_on='gene_id',
    right_index=True,
    how='left',
    suffixes=['_eqtl', '_gene']
)

sig_eqtl = sig_eqtl.dropna(subset=['chrom_gene'])
sig_eqtl = sig_eqtl.assign(
    pos_gene = (sig_eqtl.start_gene.astype(int) + sig_eqtl.end_gene.astype(int)) // 2,
)
sig_eqtl['is_cis'] = sig_eqtl.eval('chrom_eqtl == chrom_gene & (start_eqtl - 2e6) <= pos_gene <= (end_eqtl + 2e6)')
sig_eqtl.head()
```

```python
sig_eqtl_celltype = pd.merge(
    eqtl_res.groupby('gene_id', as_index=False).apply(find_lod_peaks, cell_type_specific=True).reset_index(drop=True),
    gene_locs,
    left_on='gene_id',
    right_index=True,
    how='left',
    suffixes=['_eqtl', '_gene']
)
sig_eqtl_celltype = sig_eqtl_celltype.dropna(subset=['chrom_gene'])
sig_eqtl_celltype = sig_eqtl_celltype.assign(
    pos_gene = (sig_eqtl_celltype.start_gene.astype(int) + sig_eqtl_celltype.end_gene.astype(int)) // 2,
)
sig_eqtl_celltype['is_cis'] = sig_eqtl_celltype.eval('chrom_eqtl == chrom_gene & (start_eqtl - 3e6) <= pos_gene <= (end_eqtl + 3e6)')
sig_eqtl_celltype['cluster'] = sig_eqtl_celltype.eval('(hap2_x_cluster1_pval < 0.05) + (hap2_x_cluster2_pval < 0.05) * 2').map({0: 'Neither', 1: 'Sperm', 2: 'Vegetative', 3: 'Both'})
sig_eqtl_celltype = sig_eqtl_celltype.query('cluster != "Neither"')
sig_eqtl_celltype.head()
```

```python
def chrom_axes_boilerplate(chrom_sizes, figsize=(18, 5)):

    fig, axes = plt.subplots(
        figsize=figsize,
        ncols=len(chrom_sizes),
        width_ratios=list(chrom_sizes.values()),
        sharey='all',
    )
    for chrom, ax in zip(chrom_sizes, axes):
        ax.set_xlim(0, chrom_sizes[chrom])
        ax.set_xticks(np.arange(0, chrom_sizes[chrom], 1e7))
        ax.set_xticklabels([int(i // 1e6) for i in np.arange(0, chrom_sizes[chrom], 1e7)])
        ax.set_xlabel(f'Chromosome {chrom} (Mb)')

    plt.tight_layout()
    return fig, axes
```

```python
from scipy.io import mmread

def read_index(idx_fn):
    with open(idx_fn) as f:
        idx = [r.strip().split('\t')[0] for r in f.readlines()]
    return np.array(idx)


def lognormalise_exprs(mtx, alpha=0.25):
    umi_counts = mtx.sum(0)
    sf = (umi_counts / np.mean(umi_counts))
    ln_mtx = np.log2(mtx / sf + (1.0 / (4 * alpha)))
    return ln_mtx, sf


def read_mmformat(mtx_fn, barcode_fn, feature_fn, bc_filt=200, feat_filt=3):
    mm = mmread(mtx_fn).tocsr()
    bc_mask = np.array(mm.sum(0) > bc_filt).ravel()
    mm = mm[:, bc_mask]
    feat_mask = np.array(mm.sum(1) > feat_filt).ravel()
    mm = mm[feat_mask]
    mm = np.array(mm.todense()).astype(np.float32)
    ln_mm, nf = lognormalise_exprs(mm)
    barcodes = read_index(barcode_fn)[bc_mask]
    features = read_index(feature_fn)[feat_mask]
    mm = pd.DataFrame(mm, columns=barcodes, index=features)
    ln_mm = pd.DataFrame(ln_mm, columns=barcodes, index=features)
    nf = pd.Series(nf, index=barcodes)
    return mm, ln_mm, nf

snakemake.input.solo_output = 'aligned_data/diploid/col0xdb1.col0__vs__db1_genecounts/'
count_mat, exprs_mat, nf = read_mmformat(
    os.path.join(snakemake.input.solo_output, 'matrix.mtx'),
    os.path.join(snakemake.input.solo_output, 'barcodes.tsv'),
    os.path.join(snakemake.input.solo_output, 'features.tsv'),
)
```

```python
def haplotype_expression_plotter(exprs_mat, eqtls, hap_probs, gene_locs, celltype_clusters=None, bin_size=25_000):
    hap_probs = (hap_probs > 0.5).astype(int)
    celltype_clusters = celltype_clusters.idxmax(axis=1)
    def _haplotype_expression_plot(gene_id, haplotype='cis', chrom=None, pos=None, plottype='haplotype', cluster=None, clip_zero=False):

        eqtls = eqtl_res.query('gene_id == @gene_id')
        if haplotype == 'cis':
            gene_chrom, gene_start, gene_end = gene_locs.loc[gene_id]
            cis_eqtls = eqtls.query('chrom == @gene_chrom')
            e = cis_eqtls.loc[(cis_eqtls.pos - gene_start).abs().idxmin()]
        elif haplotype == 'best_hit':
            e = eqtls.loc[eqtls.lrt_pval.idxmin()]
        elif haplotype == 'specified':
            assert chrom is not None
            chrom_eqtls = eqtls.query('chrom == @chrom')
            if pos is not None:
                e = chrom_eqtls.loc[(chrom_eqtls.pos - pos).abs().idxmin()]
            else:
                # take best hit on chrom
                e = chrom_eqtls.loc[chrom_eqtls.lrt_pval.idxmin()]
        elif haplotype == 'cis_x_trans':
            gene_chrom, gene_start, gene_end = gene_locs.loc[gene_id]
            cis_eqtls = eqtls.query('chrom == @gene_chrom')
            e = cis_eqtls.loc[(cis_eqtls.pos - gene_start).abs().idxmin()]
            assert chrom is not None
            chrom_eqtls = eqtls.query('chrom == @chrom')
            if pos is not None:
                e_t = chrom_eqtls.loc[(chrom_eqtls.pos - pos).abs().idxmin()]
            else:
                # take best hit on chrom
                e_t = chrom_eqtls.loc[chrom_eqtls.lrt_pval.idxmin()]
        else:
            raise ValueError()
        cb_haplotypes = hap_probs.loc[:, pd.IndexSlice[e.chrom, str(e.pos)]].map({0: 'hap1', 1: 'hap2'})

        df = exprs_mat.loc[gene_id].to_frame()
        if clip_zero:
            df = df[df > 0]
        #df['genotype'] = df.index.map(cb_stats.geno)
        df['haplotype'] = df.index.map(cb_haplotypes)
        if haplotype == 'cis_x_trans':
            cb_haplotypes_t = hap_probs.loc[:, pd.IndexSlice[e_t.chrom, str(e_t.pos)]].map({0: 'hap1', 1: 'hap2'})
            df['trans_haplotype'] = df.index.map(cb_haplotypes_t)
        df['celltype'] = df.index.map(celltype_clusters)
        if cluster is not None:
            df = df.query('celltype == @cluster')
        #df = df.assign(
        #    haplotype_x_parent=(df.genotype == df.haplotype).map({True: 'parent2', False: 'col0'})
        #)
        if plottype == 'haplotype_x_parent':
            raise ValueError()
            ax = sns.violinplot(
                y=gene_id,
                x='genotype',
                hue='haplotype_x_parent',
                data=df,
                cut=0,
                scale='width',
                order=genotypes,
                hue_order=['col0', 'parent2']
            )
        elif plottype == 'celltype_x_haplotype':
            ax = sns.violinplot(
                y=gene_id,
                hue='haplotype',
                x='celltype',
                data=df,
                cut=0,
                scale='width',
                hue_order=['hap1', 'hap2']
                #order=['col0'] + genotypes,
            )
        elif plottype == 'haplotype':
            if haplotype != 'cis_x_trans':
                ax = sns.violinplot(
                    y=gene_id,
                    x='haplotype',
                    data=df,
                    cut=0,
                    scale='width',
                    order=['hap1', 'hap2']
                    #order=['col0'] + genotypes,
                )
            else:
                ax = sns.violinplot(
                    y=gene_id,
                    x='haplotype',
                    hue='trans_haplotype',
                    data=df,
                    cut=0,
                    scale='width',
                    order=['hap1', 'hap2'],
                    hue_order=['hap1', 'hap2']
                    #order=['col0'] + genotypes,
                )
        ax.set_ylabel(f'{gene_id} log2 gene expression')
        title = {
            'cis': f'{gene_id} expression by haplotype\nat {gene_id} (cis-haplotype)',
            'best_hit': f'{gene_id} expression by haplotype\nat {e.chrom}:{e.pos}-{e.pos + bin_size}',
            'specified': f'{gene_id} expression by haplotype\nat {e.chrom}:{e.pos}-{e.pos + bin_size}',
            'cis_x_trans': f'{gene_id} expression by haplotype\nat {gene_id} (cis-haplotype)',
        }[haplotype]
        ax.set_title(title)
        ax.set_ylabel('Gene expression\nlog10 counts per 10k')
        return ax
    return _haplotype_expression_plot
```

```python
def region_is_syntenic(syri_file, chrom, start, end, slop=1000, min_indel_size=100):
    syn_type = 'syntenic'
    n_snvs = 0
    for record in syri_file.fetch(chrom, max(0, start - slop), end + slop):
        sv_chrom, r_start, r_end, _, _, _, q_start, q_end, _, _, sv_type, _ = record.split()
        if sv_type in ('NOTAL', 'HDR', 'DUPAL', 'INVAL', 'TRANSAL', 'INVDPAL', 'INVTRAL', 'CPG', 'CPL'):
            syn_type = 'non_syntenic'
            break
        elif sv_type == 'DEL':
            r_start, r_end = int(r_start), int(r_end)
            if (r_end - r_start) > min_indel_size:
                syn_type = 'large_indel'
        elif sv_type == 'INS':
            q_start, q_end = int(q_start), int(q_end)
            if (q_end - q_start) > min_indel_size:
                syn_type = 'large_indel'
        elif sv_type == 'SNP':
            n_snvs += 1
    return syn_type, n_snvs / (end - start) * 1000
        
@contextmanager
def synteny_mapper(syri_fns):
    syri_files = [pysam.TabixFile(syri_fn) for syri_fn in syri_fns]
    try:
        def _syn_mapper(*args, **kwargs):
            return [region_is_syntenic(sf, *args, **kwargs) for sf in syri_files]
        yield _syn_mapper
    finally:
        for sf in syri_files:
            sf.close()


def sv_mapper(gene_ids, gene_locs, syri_fns, genotypes, sig_eqtl):
    sv_res = []
    eqtl_types = []
    with synteny_mapper(syri_fns) as syn_mapper:
        for gene_id in gene_ids:
            chrom, start, end = gene_locs.loc[gene_id]
            if chrom in ('Mt', 'Pt'):
                continue
            synteny_classes = syn_mapper(chrom, start, end)
            syn_class = 'syntenic'
            n_snps = 0
            for s, n in synteny_classes:
                if s == 'non_syntenic':
                    syn_class = s
                    break
                elif s == 'large_indel':
                    syn_class = s
                n_snps += n
            n_snps = n_snps / len(syri_fns)
            gene_eqtls = sig_eqtl.query('gene_id == @gene_id')
            if not len(gene_eqtls):
                eqtl_type = 'no_eqtl'
            elif any(gene_eqtls.is_cis):
                eqtl_type = 'cis'
            else:
                eqtl_type = 'trans'
            sv_res.append([gene_id, chrom, start, end, eqtl_type, n_snps, syn_class])
    return pd.DataFrame(sv_res, columns=['gene_id', 'chrom', 'start', 'end', 'eqtl_type', 'snp_density', 'synteny_type'])
```

```python
def eqtl_plotter(gene_id, title=None, mapper=None):
    def _fdr_name_mapper(fdr_col_name):
        if mapper is None:
            if fdr_col_name == 'lrt_pval':
                return 'Combined'
            elif fdr_col_name == 'cluster_lrt_pval':
                return 'Combined celltype-specific'
            else:
                return fdr_col_name.rsplit('_', 1)[0].capitalize()
        else:
            try:
                return mapper[fdr_col_name]
            except KeyError:
                return None

    gene_eqtl_res = (eqtl_res.query('gene_id == @gene_id')
                             .melt(id_vars=['chrom', 'pos'],
                                   value_vars=[c for c in eqtl_res.columns if c.endswith('_pval')],
                                   var_name='genotype',
                                   value_name='fdr'))
    if not len(gene_eqtl_res):
        raise ValueError('Gene was not tested')
    gene_eqtl_res = gene_eqtl_res.query('~genotype.str.startswith("cluster_vs_nocluster")').assign(
        genotype=gene_eqtl_res.genotype.map(_fdr_name_mapper),
        nlog10_fdr=np.negative(np.log10(gene_eqtl_res.fdr))
    )
    if mapper is None:
        order = gene_eqtl_res.query('genotype != "Combined"').genotype.unique()
        order = ['Combined'] + sorted(order)
    else:
        order = list(mapper.values())

    gene_pos = gene_locs.loc[gene_id]

    fig, axes = chrom_axes_boilerplate(chrom_sizes, figsize=(12, 3.5))
    for (chrom, chrom_eqtl_res), ax in zip(gene_eqtl_res.groupby('chrom'), axes):
        sns.lineplot(
            x='pos',
            y='nlog10_fdr',
            hue='genotype',
            data=chrom_eqtl_res,
            drawstyle='steps-post',
            ax=ax,
            hue_order=order,
            palette = ['#252525'] + pal[:len(order) - 1]
        )
        if chrom == gene_pos.chrom:
            ax.axvline((gene_pos.start + gene_pos.end) // 2, ls='-', color='#252525', zorder=-1)
        for _, se in sig_eqtl_celltype.query('gene_id == @gene_id & chrom_eqtl == @chrom').iterrows():
            ax.axvline(se.pos_eqtl, ls='--', color='#777777', zorder=-1)
            ax.axvspan(se.start_eqtl, se.end_eqtl, color='#777777', zorder=-1, alpha=0.2)

    axes[0].set_ylabel('Negative log10 FDR')
    for ax in axes[:-1]:
        ax.legend_.remove()

    if title is None:
        title = gene_id
    fig.suptitle(title)

    plt.tight_layout(pad=0.2)
    return axes
```
