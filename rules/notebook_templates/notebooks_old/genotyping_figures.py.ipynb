```python
def parse_snp_sets(vcf_fn, minq=10):
    with pysam.VariantFile(vcf_fn) as vcf:
        parsed_snps = defaultdict(set)
        for snp in vcf.fetch():
            for acc, var in snp.samples.items():
                alt = var.alleles[0]
                parsed_snps[acc].add((snp.contig, snp.pos, alt))
        return parsed_snps


def sets_to_upset_data(sets, cat_order=None):
    if cat_order is None:
        cat_order = list(sets.keys())
    union = set()
    for set_ in sets.values():
        union.update(set_)
    union = list(union)
    memberships = []
    for item in union:
        m = []
        for label in cat_order:
            m.append(item in sets[label])
        memberships.append(m)
    memberships = pd.DataFrame(memberships, columns=cat_order)
    udata = memberships.groupby(memberships.columns.tolist()).size()
    return udata
```

```python
snps = parse_snp_sets(snakemake.input.vcf)

plot_upset(
    sets_to_upset_data(snps, cat_order=genotypes[::-1]),
    facecolor='#0072b2', sort_by='degree', sort_categories_by=None, element_size=100,
    intersection_plot_elements=4, totals_plot_elements=1, show_counts=True
)
plt.savefig(os.path.join(snakemake.output[0], 'marker_sharing_upset.svg'))
plt.show()
```
