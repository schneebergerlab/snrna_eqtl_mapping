```python
def calculate_cm_per_mb(hap_probs, nboots=100, bin_size=25_000, cm_bin_size=1_000_000, min_prob=0.01):

    nf = 1_000_000 // bin_size
    ws = cm_bin_size // bin_size
    mw = ws // 2
    
    boots = {}
    for chrom, chrom_hap_probs in hap_probs.groupby(level=0, axis=1, group_keys=False):
        chrom_hap_probs.columns = chrom_hap_probs.columns.droplevel(0)
        chrom_co_probs = (chrom_hap_probs.diff(1, axis=1)
                                         .abs()
                                         .fillna(0))
        #for i in range(int(ws)):
        #    chrom_co_probs.iloc[:, i] = chrom_co_probs.iloc[:, int(ws)]
        #    chrom_co_probs.iloc[:, -i] = chrom_co_probs.iloc[:, -int(ws)]
        chrom_co_probs.mask(chrom_co_probs < min_prob, other=0, inplace=True)
        # equivalent to rolling sum accounting for edge effects
        chrom_co_probs = chrom_co_probs.rolling(ws, center=True, min_periods=mw, axis=1).mean() * nf
        chrom_boots = []
        for _ in range(nboots):
            cm_samp = chrom_co_probs.sample(len(chrom_co_probs), replace=True).mean(axis=0)
            chrom_boots.append(cm_samp * 100)
        boots[chrom] = pd.concat(chrom_boots, axis=1)
    return boots
```

```python
def intersect(inv_a, inv_b):
    a_start, a_end = inv_a
    b_start, b_end = inv_b
    if a_end < b_start or a_start > b_end:
        return 0
    else:
        s = max(a_start, b_start)
        e = min(a_end, b_end)
        return e - s


def calculate_cM(co_invs, marker_ranges, window_size=1e6, window_step=5e4):
    chrom_sizes = co_invs.groupby('chrom', sort=True).end.max().astype(int).to_dict()
    bin_co_counts = Counter()
    window_norm_factors = Counter()
    for chrom, chrom_co_invs in co_invs.groupby('chrom'):
        chrom_bin_pos = np.arange(0, chrom_sizes[chrom] - int(window_size), int(window_step), dtype='int32')
        chrom_marker_ranges = marker_ranges.query('chrom == @chrom')
        chrom_bin_itree = NCLS(
            chrom_bin_pos,
            chrom_bin_pos + int(window_size),
            chrom_bin_pos + window_size // 2
        )
        for _, r in chrom_marker_ranges.iterrows():
            for m, n, bin_pos in chrom_bin_itree.find_overlap(
                    r.min_marker_pos, r.max_marker_pos):
                window_norm_factors[(chrom, bin_pos)] += 1
        for _, cb_co_invs in chrom_co_invs.groupby('cb'):
            if len(cb_co_invs) == 1:
                # no co on this chrom
                continue
            cb_co_invs = cb_co_invs.sort_values('start')
            co_window_starts = cb_co_invs.left_marker.values[1:]
            co_window_ends = cb_co_invs.right_marker.values[:-1]
            for i, j in zip(co_window_starts, co_window_ends):
                w = j - i
                for m, n, bin_pos in chrom_bin_itree.find_overlap(i, j):
                    bin_co_counts[(chrom, bin_pos)] += intersect((i, j), (m, n)) / w
    bin_cm_per_mb = {
        k: ((v / (window_norm_factors[k])) * 100) / (window_size / 1e6) 
        for k, v in bin_co_counts.items()
    }
    bin_cm_per_mb = pd.Series(
        list(bin_cm_per_mb.values()),
        index=pd.MultiIndex.from_tuples(bin_cm_per_mb.keys(), names=['chrom', 'pos']),
        name='cm_per_mb'
    ).reset_index()
    window_cov = pd.Series(
        list(window_norm_factors.values()),
        index=pd.MultiIndex.from_tuples(window_norm_factors.keys(), names=['chrom', 'pos']),
        name='window_coverage'
    ).reset_index()
    return bin_cm_per_mb, window_cov
```

```python
cb_marker_range = {}

for cb, cb_co_markers in co_markers.groupby('cb'):
    try:
        cb_geno = cb_genotypes[cb]
    except KeyError:
        continue
    cb_marker_range[cb] = cb_co_markers.groupby('chrom').agg({'pos': (min, max)}).assign(geno=cb_geno)

cb_marker_range = pd.concat(cb_marker_range, axis=0, names='cb').reset_index()
cb_marker_range.columns = ['cb', 'chrom', 'min_marker_pos', 'max_marker_pos', 'geno']


NBOOTS = 100
MIN_NMARKERS = np.log10(250)

order = sorted(genotypes)

cm_bins = {}

for geno in order:
    print(geno)
    geno_co_invs = co_invs.loc[co_invs.cb.isin(cb_stats.query('nmarkers >= @MIN_NMARKERS').cb)].query('geno==@geno').groupby('cb')
    geno_cb_marker_range = cb_marker_range.query('geno==@geno').groupby('cb')
    geno_cb = list(geno_co_invs.groups.keys())
    cm_bins[geno] = []
    for _ in range(NBOOTS):
        cb_sample = np.random.choice(geno_cb, replace=True, size=len(geno_cb))
        geno_co_invs_sample = pd.concat([geno_co_invs.get_group(cb) for cb in cb_sample])
        geno_cb_marker_range_sample = pd.concat([geno_cb_marker_range.get_group(cb) for cb in cb_sample])
        c, w = calculate_cM(geno_co_invs_sample, geno_cb_marker_range_sample, window_size=1e6)
        cm_bins[geno].append(c)
    cm_bins[geno] = pd.concat(cm_bins[geno])
```
