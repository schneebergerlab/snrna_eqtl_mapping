```python
cb_stats = pd.read_csv(
    snakemake.input.cb_stats,
    header=0,
    index_col=0,
    sep='\t',
).query('cluster == "hq"')

cb_whitelist = sorted(cb_stats.index.tolist())

exprs_mat, _ = lognormalise_exprs(read_mmformat(
    snakemake.input.exprs,
    cb_whitelist
))

hap_probs = pd.concat((
    pd.read_csv(fn, sep='\t', header=[0, 1], index_col=[0, 1])
    for fn in snakemake.input.crossovers
), axis=0)
hap_probs.index = hap_probs.index.droplevel('fn_idx')
hap_probs.columns = hap_probs.columns.set_levels(hap_probs.columns.levels[1].astype(int), level=1)

eqtl_res = pd.read_csv(snakemake.input.eqtls.replace('.eqtls', '.all.eqtls'), sep='\t', dtype={'chrom': str})
chrom_sizes = eqtl_res.groupby('chrom').pos.max().to_dict()
sig_eqtl = call_lod_peaks(eqtl_res, snakemake.input.gtf)
eqtl_res.head()
```

```python
pca = PCA(n_components=10, svd_solver='arpack')
pca_proj = pca.fit_transform(exprs_mat.T)
pca_proj = pd.DataFrame(pca_proj, index=exprs_mat.columns, columns=[f'PC{i}' for i in range(1, 11)])

cluster_labels = pd.Series(GaussianMixture(covariance_type='spherical', n_components=2, random_state=101).fit(pca_proj).predict(pca_proj), index=pca_proj.index)
cb_stats['celltype_cluster'] = cluster_labels
```

```python
fig, axes = chrom_axes_boilerplate(chrom_sizes, figsize=(10, 10), nrows='nchroms')
axes = axes[::-1]
chroms = list(chrom_sizes)

CIS_RANGE = 2e6
TRANS_RANGE = np.array([[25e6, 27e6]])

for i, j in it.product(range(len(chrom_sizes)), repeat=2):
    c_i, c_j = chroms[i], chroms[j]
    d = sig_eqtl.query('chrom_gene == @c_i & chrom_eqtl == @c_j')
    sns.scatterplot(
        x=d.pos_eqtl,
        y=d.pos_gene,
        hue=d.is_cis,
        size=d.lod_score,
        sizes=(5, 100),
        size_norm=(3, 50),
        hue_order=[False, True],
        palette=pal[:2],
        ax=axes[i][j],
    )
    axes[i][j].errorbar(
        x=d.pos_eqtl,
        y=d.pos_gene,
        xerr=[d.pos_eqtl - d.start_eqtl,
              d.end_eqtl - d.pos_eqtl],
        color='#777777',
        zorder=0,
        linestyle='None',
        marker='None',
        alpha=0.75
    )
    try:
        axes[i][j].legend_.remove()
    except:
        pass

    if i == j:
        axes[i][j].fill_between(
            x=[0, CIS_RANGE, chrom_sizes[c_j] - CIS_RANGE, chrom_sizes[c_j]],
            y1=[0, 0, chrom_sizes[c_i] - CIS_RANGE * 2, chrom_sizes[c_i] - CIS_RANGE],
            y2=[CIS_RANGE, CIS_RANGE * 2, chrom_sizes[c_i], chrom_sizes[c_i]],
            color=pal[1],
            alpha=0.1,
            zorder=2
        )
    
    axes[i][j].set_ylim(-500_000, chrom_sizes[c_i] + 500_000)
    axes[i][j].set_xlim(-500_000, chrom_sizes[c_j] + 500_000)

    if c_j == 'Chr1':
        for TR in TRANS_RANGE:
            if c_i != c_j:
                axes[i][j].axvspan(*TR, color=pal[2], alpha=0.2)
            else:
                axes[i][j].fill_between(
                    x=TR,
                    y1=[0, 0],
                    y2=TR - CIS_RANGE,
                    color=pal[2],
                    alpha=0.2
                )
                axes[i][j].fill_between(
                    x=TR,
                    y1=TR + CIS_RANGE,
                    y2=[chrom_sizes[c_i], chrom_sizes[c_i]],
                    color=pal[2],
                    alpha=0.2
                )

for ax, chrom in zip(axes[0], chroms):
    ax.set_xticks(np.arange(0, chrom_sizes[chrom], 1e7))
    ax.set_xticklabels([int(i // 1e6) for i in np.arange(0, chrom_sizes[chrom], 1e7)])
    ax.set_xlabel(f'{chrom} (Mb)')

for ax, chrom in zip(axes[:, 0], chroms):
    ax.set_yticks(np.arange(0, chrom_sizes[chrom], 1e7))
    ax.set_yticklabels([int(i // 1e6) for i in np.arange(0, chrom_sizes[chrom], 1e7)])
    ax.set_ylabel(f'{chrom} (Mb)')

fig.supxlabel('eQTL position')
fig.supylabel('Gene position')

plt.tight_layout(h_pad=0.1, w_pad=0.1)

plt.savefig(os.path.join(snakemake.output[0], 'eqtl_scatterplot.svg'))
plt.show()
```

```python
def region_is_syntenic(syri_file, chrom, start, end, slop=1000, min_indel_size=100):
    syn_type = 'syntenic'
    n_snvs = 0
    for record in syri_file.fetch(chrom, max(0, start - slop), end + slop):
        sv_chrom, r_start, r_end, _, _, _, q_start, q_end, _, _, sv_type, _ = record.split()
        if sv_type in ('NOTAL', 'HDR', 'DUPAL', 'INVAL', 'TRANSAL', 'INVDPAL', 'INVTRAL', 'CPG', 'CPL'):
            syn_type = 'non_syntenic'
            break
        elif sv_type == 'DEL':
            r_start, r_end = int(r_start), int(r_end)
            if (r_end - r_start) > min_indel_size:
                syn_type = 'large_indel'
        elif sv_type == 'INS':
            q_start, q_end = int(q_start), int(q_end)
            if (q_end - q_start) > min_indel_size:
                syn_type = 'large_indel'
        elif sv_type == 'SNP':
            n_snvs += 1
    return syn_type, n_snvs / (end - start) * 1000
        
@contextmanager
def synteny_mapper(syri_fns):
    syri_files = [pysam.TabixFile(syri_fn) for syri_fn in syri_fns]
    try:
        def _syn_mapper(*args, **kwargs):
            return [region_is_syntenic(sf, *args, **kwargs) for sf in syri_files]
        yield _syn_mapper
    finally:
        for sf in syri_files:
            sf.close()


def sv_mapper(gene_ids, gene_locs, syri_fns, genotypes, sig_eqtl):
    sv_res = []
    eqtl_types = []
    with synteny_mapper(syri_fns) as syn_mapper:
        for gene_id in gene_ids:
            try:
                chrom, strand, invs = gene_locs[gene_id]
            except KeyError:
                continue
            start, end = invs[0]
            if chrom in ('Mt', 'Pt'):
                continue
            synteny_classes = syn_mapper(chrom, start, end)
            syn_class = 'syntenic'
            n_snps = 0
            for s, n in synteny_classes:
                if s == 'non_syntenic':
                    syn_class = s
                    break
                elif s == 'large_indel':
                    syn_class = s
                n_snps += n
            n_snps = n_snps / len(syri_fns)
            gene_eqtls = sig_eqtl.query('gene_id == @gene_id')
            if not len(gene_eqtls):
                eqtl_type = 'no_eqtl'
            elif any(gene_eqtls.is_cis):
                eqtl_type = 'cis'
            else:
                eqtl_type = 'trans'
            sv_res.append([gene_id, chrom, start, end, eqtl_type, n_snps, syn_class])
    return pd.DataFrame(sv_res, columns=['gene_id', 'chrom', 'start', 'end', 'eqtl_type', 'snp_density', 'synteny_type'])
```
