```python
import sys
import os
import re
from glob import glob
import random
import itertools as it
import warnings
from collections import Counter

import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
import statsmodels.api as sm
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN, KMeans

sys.path.extend([os.path.abspath('../scripts/')])

from singlecell_snptools.exprs import read_mmformat, get_coverage, get_n_expressed

## Default plotting params

%matplotlib inline
sns.set(font='Arial')
plt.rcParams['svg.fonttype'] = 'none'
style = sns.axes_style('white')
style.update(sns.axes_style('ticks'))
style['xtick.major.size'] = 1
style['ytick.major.size'] = 1
sns.set(font_scale=1, style=style)
pal = sns.color_palette(['#0072b2', '#d55e00', '#009e73', '#f0e442', '#cc79a7', '#56b4e9', '#e69f00'])
cmap = ListedColormap(pal.as_hex())
sns.set_palette(pal)
sns.palplot(pal)
plt.show()
```

```python
if not os.path.exists(snakemake.output[0]):
    os.mkdir(snakemake.output[0])
```

```python
exprs_mat = read_mmformat(
    os.path.join(snakemake.input.solo_output, 'Gene/raw/matrix.mtx'),
    os.path.join(snakemake.input.solo_output, 'Gene/raw/barcodes.tsv'),
    os.path.join(snakemake.input.solo_output, 'Gene/raw/features.tsv'),
)

cb_stats = pd.read_csv(
    snakemake.input.cb_stats,
    sep='\t',
    index_col='cb'
)

cb_stats['exprs_coverage'] = cb_stats.index.map(get_coverage(
    os.path.join(snakemake.input.solo_output, 'Gene/raw/matrix.mtx'),
    os.path.join(snakemake.input.solo_output, 'Gene/raw/barcodes.tsv'),
))

cb_stats['n_expressed'] = cb_stats.index.map(get_n_expressed(
    os.path.join(snakemake.input.solo_output, 'Gene/raw/matrix.mtx'),
    os.path.join(snakemake.input.solo_output, 'Gene/raw/barcodes.tsv'),
))

exprs_mat = exprs_mat.loc[:, cb_stats.query('cluster == "hq"').index]
exprs_mat.shape
```

```python
pca_proj = PCA(n_components=5).fit_transform(exprs_mat.T)
pca_proj = pd.DataFrame(pca_proj, index=exprs_mat.columns, columns=[f'PC{i}' for i in range(1, 6)])

cluster_labels = pd.Series(KMeans(n_clusters=2).fit_predict(pca_proj), index=pca_proj.index)
cb_stats['celltype_cluster'] = cluster_labels
clusters, cluster_counts = np.unique(cluster_labels, return_counts=True)


sns.jointplot(
    x=pca_proj['PC1'],
    y=pca_proj['PC2'],
    hue=cluster_labels.map(lambda x: f'Cluster {x} (n={cluster_counts[x]})'),
    marginal_kws={'common_norm': False},
    joint_kws={'rasterized': True}
)

plt.tight_layout()
plt.savefig(os.path.join(snakemake.output[0], 'pca_cluster_jointplot.svg'), dpi=500)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 5))

norm = plt.Normalize(7, 15)
smap = plt.cm.ScalarMappable(cmap="cividis_r", norm=norm)
smap.set_array([])

sns.scatterplot(
    x=pca_proj['PC1'],
    y=pca_proj['PC2'],
    hue=cb_stats.exprs_coverage,
    palette='cividis_r',
    hue_norm=(7, 15),
    ax=ax,
    rasterized=True
)
ax.legend_.remove()
plt.colorbar(smap, ax=ax, label='log2 counts')
plt.tight_layout()
plt.savefig(os.path.join(snakemake.output[0], 'pca_sequencing_depth.svg'), dpi=500)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

sns.regplot(
    x=cb_stats.query('celltype_cluster == 1').exprs_coverage,
    y=cb_stats.query('celltype_cluster == 1').n_expressed,
    color=pal[-2],
    scatter_kws={'alpha': 0.2, 'zorder': 0},
    line_kws={'color': pal[0], 'label': 'Sperm'},
    ax=ax
)
sns.regplot(
    x=cb_stats.query('celltype_cluster == 0').exprs_coverage,
    y=cb_stats.query('celltype_cluster == 0').n_expressed,
    color=pal[-1],
    scatter_kws={'alpha': 0.5, 'zorder': 1},
    line_kws={'color': pal[1], 'label': 'Vegetative'},
    ax=ax
)
ax.set_xlabel('Unique UMIs per barcode (log2 scale)')
ax.set_ylabel('Genes detected (log2 scale)')
ax.legend(title='Nuclei type')
plt.show()
```

```python
m = sm.OLS(
    cb_stats.query('cluster == "hq"').n_expressed,
    sm.add_constant(cb_stats.query('cluster == "hq"')[['exprs_coverage', 'celltype_cluster']]),
    hasconst=True
).fit()
m.summary()
```
