```python
import sys
import os
import re
from glob import glob
import random
import itertools as it
import warnings
from collections import Counter

import numpy as np
import pandas as pd
from scipy import stats
from scipy.io import mmread
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
import statsmodels.api as sm
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

sys.path.extend([os.path.abspath('../scripts/')])

from singlecell_snptools.exprs import read_mmformat, lognormalise_exprs

## Default plotting params

%matplotlib inline
sns.set(font='Arial')
plt.rcParams['svg.fonttype'] = 'none'
style = sns.axes_style('white')
style.update(sns.axes_style('ticks'))
style['xtick.major.size'] = 1
style['ytick.major.size'] = 1
sns.set(font_scale=1, style=style)
pal = sns.color_palette(['#0072b2', '#d55e00', '#009e73', '#f0e442', '#cc79a7', '#56b4e9', '#e69f00'])
cmap = ListedColormap(pal.as_hex())
sns.set_palette(pal)
sns.palplot(pal)
plt.show()
```

```python
if not os.path.exists(snakemake.output[0]):
    os.mkdir(snakemake.output[0])
```

```python
cb_stats = pd.read_csv(
    snakemake.input.cb_stats,
    header=0,
    index_col=0,
    sep='\t',
    usecols=['cb', 'cluster'],
).query('cluster == "hq"')

cb_whitelist = sorted(cb_stats.index.tolist())

count_mat = read_mmformat(
    snakemake.input.exprs,
    cb_whitelist
)
exprs_mat, sf = lognormalise_exprs(count_mat)
cb_stats['sequencing_depth'] = np.log10(sf)
cb_stats['n_genes_exprs'] = np.log2((exprs_mat > 0).sum(0))
```

```python
pca = PCA(n_components=10, svd_solver='arpack')
pca_proj = pca.fit_transform(exprs_mat.T)
pca_proj = pd.DataFrame(pca_proj, index=exprs_mat.columns, columns=[f'PC{i}' for i in range(1, 11)])

cluster_labels = pd.Series(GaussianMixture(covariance_type='spherical', n_components=2, random_state=1).fit(pca_proj).predict(pca_proj), index=pca_proj.index)
cb_stats['celltype_cluster'] = cluster_labels
clusters, cluster_counts = np.unique(cluster_labels, return_counts=True)


sns.jointplot(
    x=pca_proj['PC1'],
    y=pca_proj['PC2'],
    hue=(cluster_labels > 0.5).astype(int).map(lambda x: f'Cluster {x} (n={cluster_counts[x]})'),
    marginal_kws={'common_norm': False},
    joint_kws={'rasterized': True},
    palette=pal[:2],
)

plt.tight_layout()
plt.savefig(os.path.join(snakemake.output[0], 'pca_cluster_jointplot.svg'), dpi=500)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

sns.regplot(
    x=cb_stats.query('celltype_cluster == 0 & -1.5 < sequencing_depth < 0.6').sequencing_depth,
    y=cb_stats.query('celltype_cluster == 0 & -1.5 < sequencing_depth < 0.6').n_genes_exprs,
    color=pal[-2],
    scatter_kws={'alpha': 0.2, 'zorder': 0},
    line_kws={'color': pal[0], 'label': 'Sperm'},
    ax=ax
)
sns.regplot(
    x=cb_stats.query('celltype_cluster == 1 & -1.5 < sequencing_depth < 0.6').sequencing_depth,
    y=cb_stats.query('celltype_cluster == 1 & -1.5 < sequencing_depth < 0.6').n_genes_exprs,
    color=pal[-1],
    scatter_kws={'alpha': 0.5, 'zorder': 1},
    line_kws={'color': pal[1], 'label': 'Vegetative'},
    ax=ax
)
ax.set_xlabel('Sequencing depth (log2 scale)')
ax.set_ylabel('Genes detected (log2 scale)')
ax.legend(title='Nuclei type')
#ax.set_xlim(-3, 3)
plt.savefig(os.path.join(snakemake.output[0], 'cluster_genes_detected.svg'), dpi=500)
plt.show()
```

```python
m = sm.OLS(
    cb_stats.query('cluster == "hq"').n_genes_exprs,
    sm.add_constant(cb_stats.query('cluster == "hq"')[['sequencing_depth', 'celltype_cluster']]),
    hasconst=True
).fit()
m.summary()
```
